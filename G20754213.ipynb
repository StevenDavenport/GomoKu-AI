```python
import copy
def reward(board_value,turn,coef):
    rewards = [0]
    for row in board_value:
        row_str = ''.join(row)
        if turn*5 in row_str.lower():
            rewards.append(coef*100)
        elif ' '+turn*4 in row_str.lower() or turn*4+' ' in row_str.lower() or turn*3+' '+turn in row_str.lower() or\
        turn+' '+turn*3 in row_str.lower() or turn*2+' '+turn*2 in row_str.lower():
            rewards.append(coef*100*(4/5))
        elif ' '+turn*3+' ' in row_str.lower():
             rewards.append(coef*100*(3.5/5))
        elif '  '+turn*3 in row_str.lower() or turn*3+'  ' in row_str.lower() or \
        ' '+turn*2+' '+turn in row_str.lower() or turn+' '+turn*2+' ' in row_str.lower():
            rewards.append(coef*100*(3/5))
        elif ' '*3+turn*2 in row_str.lower() or turn*2+' '*3 in row_str.lower() or ' '*2+turn*2+' '*1 in row_str.lower() or\
        ' '*1+turn*2+' '*2 in row_str.lower():
            rewards.append(coef*100*(2/5))
        
    for i in range(0,board_value.shape[1]):
        col = board_value[:,i]
        col_str = ''.join(col)
        if turn*5  in col_str.lower():
            rewards.append(coef*100)
        elif ' '+turn*4 in col_str.lower() or turn*4+' ' in col_str.lower() or turn*3+' '+turn in col_str.lower() or\
        turn+' '+turn*3 in col_str.lower() or turn*2+' '+turn*2 in col_str.lower():
            rewards.append(coef*100*(4/5))
        elif ' '+turn*3+' ' in col_str.lower():
             rewards.append(coef*100*(3.5/5))
        elif  '  '+turn*3 in col_str.lower() or turn*3+'  ' in col_str.lower() or\
        ' '+turn*2+' '+turn in col_str.lower() or turn+' '+turn*2+' ' in col_str.lower():
            rewards.append(coef*100*(3/5))
        elif ' '*3+turn*2 in col_str.lower() or turn*2+' '*3 in col_str.lower() or ' '*2+turn*2+' '*1 in col_str.lower() or\
        ' '*1+turn*2+' '*2 in col_str.lower():
            rewards.append(coef*100*(2/5))

    for i in range(0,board_value.shape[1]):
        diagonal = board_value.diagonal(offset=i)
        diagonal_str = ''.join(diagonal)
        if turn*5 in diagonal_str.lower():
            rewards.append(coef*100*(5/5))
        elif ' '+turn*4 in diagonal_str.lower() or turn*4+' ' in diagonal_str.lower() or turn*3+' '+turn in diagonal_str.lower() or\
        turn+' '+turn*3 in diagonal_str.lower() or turn*2+' '+turn*2 in diagonal_str.lower():
            rewards.append(coef*100*(4/5))
        elif ' '+turn*3+' ' in diagonal_str.lower():
             rewards.append(coef*100*(3.5/5))
        elif  '  '+turn*3 in diagonal_str.lower() or turn*3+'  ' in diagonal_str.lower() or\
        ' '+turn*2+' '+turn in diagonal_str.lower() or turn+' '+turn*2+' ' in diagonal_str.lower():
            rewards.append(coef*100*(3/5))
        elif ' '*3+turn*2 in diagonal_str.lower() or turn*2+' '*3 in diagonal_str.lower() or ' '*2+turn*2+' '*1 in diagonal_str.lower() or\
        ' '*1+turn*2+' '*2 in diagonal_str.lower():
            rewards.append(coef*100*(2/5))
    
    for i in range(0,board_value.shape[0]):
        diagonal = board_value.diagonal(offset=i,axis1=1,axis2=0)
        diagonal_str = ''.join(diagonal)
        if turn*5 in diagonal_str.lower():
            rewards.append(coef*100*(5/5))
        elif ' '+turn*4 in diagonal_str.lower() or turn*4+' ' in diagonal_str.lower() or turn*3+' '+turn in diagonal_str.lower() or\
        turn+' '+turn*3 in diagonal_str.lower() or turn*2+' '+turn*2 in diagonal_str.lower():
            rewards.append(coef*100*(4/5))
        elif ' '+turn*3+' ' in diagonal_str.lower():
             rewards.append(coef*100*(3.5/5))
        elif  '  '+turn*3 in diagonal_str.lower() or turn*3+'  ' in diagonal_str.lower() or\
        ' '+turn*2+' '+turn in diagonal_str.lower() or turn+' '+turn*2+' ' in diagonal_str.lower():
            rewards.append(coef*100*(3/5))
        elif ' '*3+turn*2 in diagonal_str.lower() or turn*2+' '*3 in diagonal_str.lower() or ' '*2+turn*2+' '*1 in diagonal_str.lower() or\
        ' '*1+turn*2+' '*2 in diagonal_str.lower():
            rewards.append(coef*100*(2/5))
    
    flipped_board_value = np.flip(board_value,axis=1)
    for i in (range(flipped_board_value.shape[0])):
        diagonal = flipped_board_value.diagonal(offset=i,axis1=0,axis2=1)
        diagonal_str = ''.join(diagonal)
        if turn*5 in diagonal_str.lower():
            rewards.append(coef*100*(5/5))
        elif ' '+turn*4 in diagonal_str.lower() or turn*4+' ' in diagonal_str.lower() or turn*3+' '+turn in diagonal_str.lower() or\
        turn+' '+turn*3 in diagonal_str.lower() or turn*2+' '+turn*2 in diagonal_str.lower():
            rewards.append(coef*100*(4/5))
        elif ' '+turn*3+' ' in diagonal_str.lower():
             rewards.append(coef*100*(3.5/5))
        elif  '  '+turn*3 in diagonal_str.lower() or turn*3+'  ' in diagonal_str.lower() or\
        ' '+turn*2+' '+turn in diagonal_str.lower() or turn+' '+turn*2+' ' in diagonal_str.lower():
            rewards.append(coef*100*(3/5))
        elif ' '*3+turn*2 in diagonal_str.lower() or turn*2+' '*3 in diagonal_str.lower() or ' '*2+turn*2+' '*1 in diagonal_str.lower() or\
        ' '*1+turn*2+' '*2 in diagonal_str.lower():
            rewards.append(coef*100*(2/5))
        
    for i in (range(flipped_board_value.shape[1])):
        diagonal = flipped_board_value.diagonal(offset=i,axis1=1,axis2=0)
        diagonal_str = ''.join(diagonal)

        if turn*5 in diagonal_str.lower():
            rewards.append(coef*100*(5/5))
        elif ' '+turn*4 in diagonal_str.lower() or turn*4+' ' in diagonal_str.lower() or turn*3+' '+turn in diagonal_str.lower() or\
        turn+' '+turn*3 in diagonal_str.lower() or turn*2+' '+turn*2 in diagonal_str.lower():
            rewards.append(coef*100*(4/5))
        elif ' '+turn*3+' ' in diagonal_str.lower():
             rewards.append(coef*100*(3.5/5))
        elif  '  '+turn*3 in diagonal_str.lower() or turn*3+'  ' in diagonal_str.lower() or\
        ' '+turn*2+' '+turn in diagonal_str.lower() or turn+' '+turn*2+' ' in diagonal_str.lower():
            rewards.append(coef*100*(3/5))
        elif ' '*3+turn*2 in diagonal_str.lower() or turn*2+' '*3 in diagonal_str.lower() or ' '*2+turn*2+' '*1 in diagonal_str.lower() or\
        ' '*1+turn*2+' '*2 in diagonal_str.lower():
            rewards.append(coef*100*(2/5))
    
    if turn=='x':
        return min(rewards)
    else:
        return max(rewards)
    
def find_local_board(board,last_move,grid_size):
    row_idx = np.where(board_idx==last_move)[0][0]
    col_idx = np.where(board_idx==last_move)[1][0]
    upper_b = row_idx
    left_b = col_idx
    bottom_b = row_idx
    right_b = col_idx
    
    if row_idx-grid_size<=0:
        upper_b = 0
    else:
        upper_b = row_idx-grid_size
    
    if col_idx-grid_size<0:
        left_b = 0
    else:
        left_b = col_idx-grid_size
        
    if row_idx+grid_size>=15:
        bottom_b = 15
    else:
        bottom_b = row_idx+grid_size+1
    
    if col_idx+grid_size>=15:
        right_b = 15
    else:
        right_b = col_idx+grid_size+1
    
    local_board_idx = board_idx[upper_b:bottom_b,left_b:right_b]
    local_board_values = board[upper_b:bottom_b,left_b:right_b]
    return local_board_idx,local_board_values

def Lis_Agent_Policy(board, last_move, my_turn=True, last_cost=-1000,depth=0):
    move_idx = np.where(board!=' ')
#     depth_limit = 2
    grid_size = 5
    
   
        
    if last_move is None:
        random_move = dummy_policy(board)
        return random_move,0
    depth = depth+1

    local_board = copy.copy(board)
    max_min_cost = 0
    max_min_node = None
    
    
    if my_turn:
        max_min_cost = -1000
    else:
        max_min_cost = 1000
    '''
    check to see if the game is terminated 
    '''
    terminal_status = terminal(local_board)
    if terminal_status == 'X':
        return max_min_node, -100
    elif terminal_status == 'O':
        return max_min_node, 100
    elif terminal_status == 'tie':
        return max_min_node, 0
    
    local_board_idx,local_board_values = find_local_board(local_board,last_move,grid_size)
    available_moves_idx = np.where(local_board_values==' ')
    avaiable_moves = local_board_idx[available_moves_idx]
    avaiable_values = local_board_values[available_moves_idx]    
    
    available_values_idx = np.where(local_board_values==' ')
    no_of_avaiable_values = len(local_board_values[available_values_idx])
    
    depth_limit = 2
    if  no_of_avaiable_values/len(local_board_values)<0.5:
        depth_limit = 3
    elif no_of_avaiable_values/len(local_board_values)<0.2:
        depth_limit = 20
    
    '''
    For all avaiable moves from the current board state and caculate costs for each of the move
    and store the node with max/min cost in the max_min_node and max_min_cost
    '''
  
    for move in avaiable_moves:
        row_idx = np.where(board_idx==move)[0][0]
        col_idx = np.where(board_idx==move)[1][0]
        
        if my_turn:
            local_board[row_idx,col_idx] = 'O'

            if depth<depth_limit:
                next_move_min, next_node_cost = Lis_Agent_Policy(local_board,last_move, my_turn=False,last_cost=max_min_cost,depth=depth)
            else:
                next_node_cost = 0
            
            node_cost = next_node_cost+reward(local_board,'o',1)
            if last_cost!=-1000 and node_cost > last_cost:
                local_board[row_idx,col_idx] = ' '
                max_min_cost = node_cost
                max_min_node = move
                break
                
            if node_cost>max_min_cost:
                max_min_cost = node_cost
                max_min_node = move
            
            local_board[row_idx,col_idx] = ' '
           
        else:
            local_board[row_idx,col_idx] = 'X'

            if depth<depth_limit:
                _, next_node_cost = Lis_Agent_Policy(local_board, last_move,my_turn=True,last_cost=max_min_cost,depth=depth)
            else:
                next_node_cost = 0
            node_cost = next_node_cost+reward(local_board,'x',-1)
            if last_cost!=-1000 and node_cost < last_cost:
                local_board[row_idx,col_idx] = ' '
                max_min_cost = node_cost
                max_min_node = move
                break

            if node_cost<max_min_cost:
                max_min_cost = node_cost
                max_min_node = move
            
            local_board[row_idx,col_idx] = ' '


    return max_min_node, max_min_cost

def expectimax_policy(board, last_move,my_turn=True, last_cost=-1000,depth=0):
    move_idx = np.where(board!=' ')
#     depth_limit = 2
    grid_size = 5
    
#     if (len(board_idx[move_idx])/255)>0.2 and (len(board_idx[move_idx])/255)<=0.8:
#         depth_limit = 3
#         grid_size = 5
#     elif (len(board_idx[move_idx])/255)>0.8:
#         depth_limit = 3
#         grid_size=5

    if last_move is None:
        random_move = dummy_policy(board)
        return random_move,0
    depth = depth+1

    local_board = copy.copy(board)
    max_min_cost = 0
    max_min_node = None
    min_reward_list = []
    
    if my_turn:
        max_min_cost = -1000
    else:
        max_min_cost = 1000
    '''
    check to see if the game is terminated 
    '''
    terminal_status = terminal(local_board)
    if terminal_status == 'X':
        return max_min_node, -100
    elif terminal_status == 'O':
        return max_min_node, 100
    elif terminal_status == 'tie':
        return max_min_node, 0
    
    local_board_idx,local_board_values = find_local_board(local_board,last_move,grid_size)
    available_moves_idx = np.where(local_board_values==' ')
    avaiable_moves = local_board_idx[available_moves_idx]
    avaiable_values = local_board_values[available_moves_idx]    
    
    available_values_idx = np.where(local_board_values==' ')
    no_of_avaiable_values = len(local_board_values[available_values_idx])
    
    depth_limit = 2
    if no_of_avaiable_values/len(local_board_values)<0.5:
        depth_limit = 3
    elif no_of_avaiable_values/len(local_board_values)<0.2:
        depth_limit = 20
    '''
    For all avaiable moves from the current board state and caculate costs for each of the move
    and store the node with max/min cost in the max_min_node and max_min_cost
'''
    for move in avaiable_moves:
        row_idx = np.where(board_idx==move)[0][0]
        col_idx = np.where(board_idx==move)[1][0]

        if my_turn:
            local_board[row_idx,col_idx] = 'O'
            
            if depth<depth_limit:
                _,next_node_cost = expectimax_policy(local_board,last_move,my_turn=False,last_cost=max_min_cost,depth=depth)
            else:
                next_node_cost = 0
                
            node_cost = next_node_cost+reward(local_board,'o',1)

            if node_cost>max_min_cost:
                max_min_cost = node_cost
                max_min_node = move
            local_board[row_idx,col_idx] = ' '
        else:
            local_board[row_idx,col_idx] = 'X'

            if depth<depth_limit:
                _, next_node_cost = expectimax_policy(local_board, last_move,my_turn=True,last_cost=max_min_cost,depth=depth)
            else:
                next_node_cost = 0
            node_cost = next_node_cost+reward(local_board,'x',-1)
            min_reward_list.append((move,node_cost))
            local_board[row_idx,col_idx] = ' '

    if my_turn:
        return max_min_node, max_min_cost
    else:
        final_node_cost = 0
        p = 1/len(avaiable_moves)
        for move_cost in min_reward_list:
            final_node_cost = move_cost[1]*p+final_node_cost
            if(move_cost[1]*p<max_min_cost):
                max_min_cost = move_cost[1]*p
                max_min_node = move_cost[0]
        return max_min_node,final_node_cost
```
